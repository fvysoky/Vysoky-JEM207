```python
import numpy as np
import pandas as pd
import random

np.random.seed(38496905)
df = pd.DataFrame()
df["customer_id"] = np.random.randint(1,40, size=200)
categories = ("electronics","clothing","books","foods")
df["category"] = np.random.choice(categories, size=200)
prices = np.random.normal(50,15,200)
while prices.any() < 0:
    prices = np.random.normal(50,15,200)
df["price"] = prices.round(2)
df["quantity"] = np.random.randint(1,5,200)
discounts = (0,0.1,0.2)
df["discount"] = np.random.choice(discounts, size=200)
df["day"] = np.random.randint(1,30,200)
df
```

```python
df["gross_revenue"] = df["price"] * df["quantity"]
df["net_revenue"] = (df["gross_revenue"] * (1 - df["discount"])).round(2)
df["category_avg_net_revenue"] = df.groupby(by = "category")["net_revenue"].transform("mean").round(2)
df[df["net_revenue"] > df["category_avg_net_revenue"]]
```

```python
df["order_type"] = df.apply(lambda row: 
                            "bulk_discounted" if (row["quantity"] >= 4) & (row["discount"] > 0) else
                            "bulk_fullprice" if (row["quantity"] >= 4) & (row["discount"] == 0) else
                            "standard_discounted" if (row["discount"]>0) & (row["quantity"]<4) else 
                            "standard_fullprice",axis=1)
sum_table = pd.DataFrame({"average_price" : df.groupby(by = "category")["price"].agg("mean").round(2), 
                        "total_quantity" : df.groupby(by = "category")["quantity"].agg("sum"),
                        "average_discount" : df.groupby(by = "category")["discount"].agg("mean").round(2),
                         "total net revenue" : df.groupby(by = "category")["net_revenue"].agg("sum"),
                         "number of transactions" : df.groupby(by = "category").size()})
sum_table
```
